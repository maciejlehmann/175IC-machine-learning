```
import pandas as pd
import numpy as np
from sklearn import preprocessing

data = {
    'size' : ['L', 'L', 'M', 'XL', 'M', 'S', 'M', 'XXL', 'XL', 'L', 'M', 'XL', 'S', 'S', 'XL', 'M', 'L', 'XXL', 'S', 'XL'],
    'color' : ['green', 'blue', 'red', 'blue', 'green', 'black', 'white', 'yellow', 'blue', 'black', 'green', 'blue', 'white', 'black', 'yellow', 'black', 'white', 'yellow', 'blue', 'black'],
    'gender' : ['male', 'female', 'male', 'female', 'male', 'female', 'male', 'female', 'male', 'female', 'male', 'female', 'male', 'female', 'male', 'female', 'male', 'female', 'male', 'female'],
    'price' : [129.0, 90.0, 111.0, 141.0, 120.0, 84.0, 97.0, 174.0, 150.0, 230.0, 84.0, 129.0, 141.0, 165.0, 191.0, 80.0, 150.0, 172.0, 233.0, 97.0],
    'weight' : [45, 67, 65, 78, 73, 89, 102, 98, 78, 77, 73, 45, 101, 73, 64, 67, 100, 91, 59, 61],
    'bought': ['no', 'yes', 'no', 'yes', 'no', 'yes', 'no', 'yes', 'no', 'yes', 'no', 'yes', 'no', 'yes', 'no', 'no', 'yes', 'no', 'yes', 'no']
}

df_raw = pd.DataFrame(data=data)
df = df_raw.copy()

#df.info()

for col in ['size', 'color', 'gender', 'bought']:
    df[col] = df[col].astype('category')

for col2 in ['weight']:
  df[col2] = df[col2].astype('float')

#df.info()

df.describe().T
df.describe(include=['category']).T

df

le = preprocessing.LabelEncoder()
df['bought'] = le.fit_transform(df['bought'])

df

pd.get_dummies(data=df)

pd.get_dummies(data=df, drop_first=True)

scaler = preprocessing.StandardScaler()
df[['price', 'weight']] = scaler.fit_transform(df[['price', 'weight']])

df
df_raw
```
