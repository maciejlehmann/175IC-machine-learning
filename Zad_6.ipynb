```
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt
from sklearn import metrics
import numpy as np

# Zadanie 1

iris = datasets.load_iris()
X = iris.data[:, :2]
y = iris.target

# Zadanie 2

x_train,x_test,y_train,y_test=train_test_split(X, y, test_size=0.4)
```

```
# Zadanie 3

clf = SVC(C=1.5, kernel='linear')
clf.fit(x_train, y_train)

print(clf)
```

```
# Zadanie 4 - C = 0.08, kernel = linear

clf = SVC(C=0.08, kernel='linear')
clf.fit(x_train, y_train)

y_pred = clf.predict(x_test)
accuracy = metrics.accuracy_score(y_test, y_pred)
plot_decision_regions(x_train, y_train, clf=clf)

plt.title(f'Dokładność klasyfikacji = {round(accuracy, 2)}')
plt.show()
```

```
# Zadanie 4 - C = 0.333, kernel = poly

clf = SVC(C=0.333, kernel='poly')
clf.fit(x_train, y_train)

y_pred = clf.predict(x_test)
accuracy = metrics.accuracy_score(y_test, y_pred)
plot_decision_regions(x_train, y_train, clf=clf)

plt.title(f'Dokładność klasyfikacji = {round(accuracy, 2)}')
plt.show()
```

```
# Zadanie 4 - C = 1, kernel = rbf

clf = SVC(C=1, kernel='rbf')
clf.fit(x_train, y_train)

y_pred = clf.predict(x_test)
accuracy = metrics.accuracy_score(y_test, y_pred)
plot_decision_regions(x_train, y_train, clf=clf)

plt.title(f'Dokładność klasyfikacji = {round(accuracy, 2)}')
plt.show()
```

```
# Zadanie 4 - C = 0.05, kernel = sigmoid

clf = SVC(C=0.7, kernel='sigmoid')
clf.fit(x_train, y_train)

y_pred = clf.predict(x_test)
accuracy = metrics.accuracy_score(y_test, y_pred)
plot_decision_regions(x_train, y_train, clf=clf)

plt.title(f'Dokładność klasyfikacji = {round(accuracy, 2)}')
plt.show()
```
