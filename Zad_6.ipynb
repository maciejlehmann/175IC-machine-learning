```
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC as svc
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt
from sklearn import metrics
import numpy as np


# Zadanie 1

raw_data = load_iris()
all_data = raw_data.copy()

input_data = all_data['data'][:, :2]
target = all_data['target']

# Zadanie 2

X_train, X_test, y_train, y_test = train_test_split(input_data, target)
```

```
# Zadanie 3

clf = svc(C=1.5, kernel='linear')
clf.fit(X_train, y_train)

print(clf)
```

```
# Zadanie 4 - C = 0.08, kernel = linear

svc = SVC(C=0.08, kernel='linear')
svc.fit(X_train, y_train)
plot_decision_regions(X_train, y_train, clf=svc, legend=2)
pred = svc.predict(X_test)
acc = svc.score(X_test, y_test)

plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.title(f"Accuracy: {round(acc, 4)}")
plt.show()
```

```
# Zadanie 4 - C = 0.333, kernel = poly

svc = SVC(C=0.333, kernel='poly')
svc.fit(X_train, y_train)
plot_decision_regions(X_train, y_train, clf=svc, legend=2)
pred = svc.predict(X_test)
acc = svc.score(X_test, y_test)

plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.title(f"Accuracy: {round(acc, 4)}")
plt.show()
```
